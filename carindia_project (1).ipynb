```python
f=data.drop(['city','fuel_type','model_name'],axis=1)
```

```python
f
```

```python
c=f.loc[(f.model_name1=="Wagon")]
```

```python
c
```

```python
d=f.loc[(f.model_name1=='Swift')]
```

```python
d
```

```python
e=f.loc[(f.model_name1=="Alto")]
```

```python
e
```

```python
t=pd.concat([c,d,e])
```

```python
t
```

```python
g=pd.get_dummies(t,dtype=int)
```

```python
g
```

```python
y=g['price (₹)']
y
```

```python
x=g.drop(['price (₹)'],axis=1)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=41)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
```

```python
reg.fit(x_train,y_train)
```

```python
ypred=reg.predict(x_test)
```

```python
ypred
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,ypred) 
```
